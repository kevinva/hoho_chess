```python
import os
import time
import json
import numpy as np
import pandas as pd
```

```python
data_dir = "../output/data"
file_list = os.listdir(data_dir)
file_count = len(file_list)
item_list = list()

for i, filename in enumerate(file_list):
    file_path = os.path.join(data_dir, filename)
    print(file_path)
    with open(file_path, "r") as file:
        json_str = file.read()
        matches = json.loads(json_str)
        match_count = len(matches)
        
        for j, match in enumerate(matches):
            print(f"match: {match}")
            break
            
            mid = i * file_count + j
            base_id =  f"{int(time.time())}"
            for k, round in enumerate(match):
                chapture_list = round[3]
                win = 0
                if "K" in chapture_list:
                    win = -1
                elif "k" in chapture_list:
                    win = 1
                    
                round_dict = {"match_id": f"{mid}",
                              "round_id": f"{base_id}{k}",
                              "state": round[0],
                              "action": round[2],
                              "chaptures": ','.join(chapture_list),
                              "h_reward": round[4],
                              "win": win}
                item_list.append(round_dict)
                
                if win == 1 or win == -1:
                    break
```

```python
data = pd.DataFrame(item_list)
```
